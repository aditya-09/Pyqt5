```python
import sys

from PyQt5.QtWidgets import (
    QApplication,QPushButton,
    QComboBox,
    QFormLayout,
    QLineEdit,
    QStackedLayout,
    QVBoxLayout,
    QWidget,
)

class Window(QWidget):
    def __init__(self):
        super().__init__()
        self.setWindowTitle("College Records For Placements")
        # Create a top-level layout
        layout = QVBoxLayout()
        self.setLayout(layout)
        
        # Create and connect the combo box to switch between pages
        self.pageCombo = QComboBox()
        list1=["Rank","Unemployment Rate", "Jobs"]
        self.pageCombo.addItems(list1)
        self.figure = plt.figure() 
        self.resize(1200,800)
        self.canvas = FigureCanvas(self.figure)
        self.toolbar = NavigationToolbar(self.canvas, self) 
   
        self.button = QPushButton('Show Data')
        self.button.clicked.connect(self.page1)
        #self.button.clicked.connect(self.page2)
        layout.addWidget(self.toolbar) 
        layout.addWidget(self.canvas) 
        layout.addWidget(self.button) 
        self.setLayout(layout) 

        self.pageCombo.activated.connect(self.switchPage)
        # Create the stacked layout
        self.stackedLayout = QStackedLayout()
        # Create the first page
        self.page1 = QWidget()
        self.page1Layout = QFormLayout()
        self.page1.setLayout(self.page1Layout)
        self.stackedLayout.addWidget(self.page1)
        
        # Create the second page
        self.page2 = QWidget()
        self.page2Layout = QFormLayout()
        self.page2.setLayout(self.page2Layout)
        self.stackedLayout.addWidget(self.page2)
        
        # Add the combo box and the stacked layout to the top-level layout
        layout.addWidget(self.pageCombo)
        layout.addLayout(self.stackedLayout)

   
    def page1(self):
        # random data 
        index = self.pageCombo.currentIndex() 
  
        # showing content on the screen though label 
        if index==0:
            data ='https://raw.githubusercontent.com/fivethirtyeight/data/master/college-majors/recent-grads.csv'
            df = pd.read_csv(data,index_col=0)
            df.insert(0, 'College Rank', range(1, 1 + len(df)))
        
            # clearing old figure 
            self.figure.clear() 
   
            # create an axis 
            ax0 = self.figure.add_subplot(111) 
   
            # plot data 
            plt.title('College Rank & Full Time')
            plt.xlabel('College_Rank')
            plt.ylabel('Full_Time_Students')
            plt.plot(df['College Rank'], df['Full_time'],linewidth=1) 
            plt.show()
            # refresh canvas 
            self.canvas.draw()
        
        elif index==1:
            data ='https://raw.githubusercontent.com/fivethirtyeight/data/master/college-majors/recent-grads.csv'
            df = pd.read_csv(data,index_col=0)
            df.insert(0, 'College Rank', range(1, 1 + len(df)))
        
            # clearing old figure 
            self.figure.clear() 
   
            # create an axis 
            ax0 = self.figure.add_subplot(111) 
   
            # plot data 
            plt.title('Unemployment Ratio')
            plt.xlabel('Part Time Students')
            plt.ylabel('Unemployment Rate')
            plt.plot(df['Part_time'], df['Unemployment_rate'],color='r')
            plt.show()
            # refresh canvas 
            self.canvas.draw()
            
        elif index==2:
            data ='https://raw.githubusercontent.com/fivethirtyeight/data/master/college-majors/recent-grads.csv'
            df = pd.read_csv(data,index_col=0)
            df.insert(0, 'College Rank', range(1, 1 + len(df)))
        
            # clearing old figure 
            self.figure.clear() 
   
            # create an axis 
            ax0 = self.figure.add_subplot(111) 
   
            # plot data 
            plt.bar(df['College Rank'], df['College_jobs'], color ='g', label ='College jobs') 
            plt.bar(df['College Rank'], df['Non_college_jobs'], color ='r', label ='Non College jobs')
            plt.xlabel('College Rank', fontweight ='bold') 
            plt.ylabel('Jobs', fontweight ='bold') 
            plt.show() 
   
            # refresh canvas 
            self.canvas.draw()
    def switchPage(self):
        self.stackedLayout.setCurrentIndex(self.pageCombo.currentIndex())
if __name__ == "__main__":
    app = QApplication(sys.argv)
    window = Window()
    window.show()
    sys.exit(app.exec_())
```
